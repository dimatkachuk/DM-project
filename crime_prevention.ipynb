```r
crimes <- read.csv('crimes.csv')
```

```r
d <- crimes$Date[6:10]
d
```

```r
as.POSIXlt(d, format = "%m/%d/%Y %I:%M:%S %p")
```

```r
#Add column hour of the day
crimes$Hour <- format(as.POSIXlt(crimes$Date, format = "%m/%d/%Y %I:%M:%S %p"),"%H")
```

```r
#ADDING WEEKDAYS
crimes$weekday <- weekdays(as.Date(crimes$Date, "%m/%d/%Y"))
```

```r
#without time
crimes$Date <- format(as.Date(crimes$Date, format = "%m/%d/%Y"),"%m/%d/%Y")
```

```r
#Adding federal holidays

federal_holiday = read.csv('federal_holiday.csv')
federal_holiday$DTSTART <- format(as.Date(federal_holiday$DTSTART, format = "%m/%d/%Y"),"%m/%d/%Y")

hol_list <- as.vector(federal_holiday$DTSTART)
```

```r
hol_list
```

```r

#Chicago day-offs
listCH <- which (crimes$Date %in% hol_list)
crimes$holiday <- 0
for (i in listCH){
crimes$holiday[listCH] <- TRUE
}
```

```r
table(crimes$holiday)
```
